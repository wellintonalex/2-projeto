### Destaques

* **Count:** Indica o número total de respostas válidas para cada pergunta (neste caso, 23 para cada coluna).
* **Mean (Média):** Representa a média aritmética dos valores para cada coluna. No caso de "Quantas porções de frutas você consome em uma semana?", a média é de aproximadamente 4,48 porções, o que sugere um consumo moderado. Já a média de 10,39 para bebidas açucaradas indica um consumo elevado em comparação com frutas.
* **Std (Desvio Padrão):** Mede a variabilidade dos dados. Um valor alto de desvio padrão, como 12,61 para bebidas açucaradas e 14,04 para horas de dispositivos móveis, indica que há uma grande variação nas respostas entre os participantes.
* **Min e Max:** Mostram os valores mínimo e máximo para cada coluna. Por exemplo, o consumo de frutas varia de 0 até 40 porções por semana, enquanto o tempo em dispositivos móveis vai de 2 a 50 horas semanais.

```python
mais_frequente_horario = dados["melhor horario"].mode()[0]
print(f"O horário mais frequente para treinar é: {mais_frequente_horario}")
```

```python
mais_frequente_dia_peito = dados["melhor dia para treinar peito"].mode()[0]
print(f"O dia mais frequente para treinar peito é: {mais_frequente_dia_peito}")
```
